```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

<stronga>1.Normal Distribution:</strong><small> Data is symmetrically distributed with no skew. Most of the data falls in the middle and tapers off on either side.</samll>

```python
normal_data = np.random.normal(size=1000)
plt.figure(figsize=(10,6))
sns.histplot(normal_data, kde=True)
plt.title('Normal Distribution')
plt.show()
```

<strong>Right-skewed Distribution (Positive Skew):</strong> <small>Most of the data is concentrated on the left side, with a tail extending to the right.</small>

```python
right_skewed_data = np.random.exponential(size=1000)
plt.figure(figsize=(10,6))
sns.histplot(right_skewed_data, kde=True)
plt.title('Right-skewed Distribution')
plt.show()
```

<strong>Left-skewed Distribution (Negative Skew):</strong> <small> Most of the data is concentrated on the right side, with a tail extending to the left.</small>

```python
left_skewed_data = -np.random.exponential(size=1000)
plt.figure(figsize=(10,6))
sns.histplot(left_skewed_data, kde=True)
plt.title('Left-skewed Distribution')
plt.show()
```

<strong>Uniform Distribution:</strong> <small>Data is evenly distributed across the range. Each value has an equal chance of being chosen.</small>

```python
uniform_data = np.random.uniform(size=1000)
plt.figure(figsize=(10,6))
sns.histplot(uniform_data, kde=True)
plt.title('Uniform Distribution')
plt.show()
```

<strong>Bimodal Distribution:</strong> <small> Data has two different modes. This can happen if the data set is a mix of two separate groups.</small>

```python
bimodal_data = np.concatenate([np.random.normal(-1, 1, size=500), np.random.normal(2, 1, size=500)])
plt.figure(figsize=(10,6))
sns.histplot(bimodal_data, kde=True)
plt.title('Bimodal Distribution')
plt.show()
```
