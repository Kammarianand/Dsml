<strong>Heart Disease EDA</strong>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
hrt_data = pd.read_csv("heart.csv")

```

```python
hrt_data.shape
```

```python
hrt_data.head(10)
```

<strong>Analyzing numerical values using dis plot</strong>

<p>displot is a function in Seaborn library in Python that provides a flexible interface for drawing distribution plots.</p>

```python
sns.displot(hrt_data['Age'],kde=True)
```

```python
sns.distplot(hrt_data["RestingBP"],kde=True)
```

```python
sns.distplot(hrt_data["Cholesterol"],kde=True)
```

<strong>Categorical variable analysis using pie charts</strong>

<u>The groupby method in pandas is used to split the data into groups based on some criteria, allowing complex data operations to be performed on these groups.</u>

```python
hrt_data.groupby('Sex').size().plot(kind="pie",autopct="%.1f")
plt.title("Pie chart")
```

```python
hrt_data.groupby('ChestPainType').size().plot(kind="pie",autopct="%.1f")
```

```python
hrt_data.groupby('RestingECG').size().plot(kind="pie",autopct="%.1f")
plt.axis("off")
```

```python
hrt_data.groupby('ST_Slope').size().plot(kind="pie",autopct="%.1f")
```

```python
# HeartDisease
hrt_data.groupby('HeartDisease').size().plot(kind="pie",autopct="%.1f")
```

<strong>Violin plot</strong>: &nbsp;
<small>A violin plot is a method of plotting numeric data that combines a box plot and a kernel density plot, providing a visual representation of both the distribution and summary statistics of the data.</small>

```python
sns.violinplot(x = hrt_data['HeartDisease'],y = hrt_data['Age'],hue=hrt_data["Sex"])
```

<strong>correlation analysis</strong>: &nbsp;
<small>Correlation analysis is a statistical method used to evaluate the strength and direction of the relationship between two quantitative variables.</small>

```python
corr = hrt_data.corr()
corr
```

```python
sns.heatmap(corr)
```

<strong>jointplot:</strong>&nbsp;<small> A joint plot in Seaborn is a method of plotting that provides a 2D visualization of the relationship between two variables, along with their individual distributions as histograms or KDE plots on the same chart.</small>


```python
sns.jointplot(x = "Age",y = "MaxHR",data = hrt_data,kind="reg")
```

```python
sns.jointplot(x = 'Cholesterol',y="MaxHR",data=hrt_data,kind="reg")
```

<strong>PairPlot:</strong>&nbsp;<small> A pair plot is a type of plot that shows the pairwise relationships between the variables in a dataset. </small>

```python
sns.pairplot(hrt_data)
```
